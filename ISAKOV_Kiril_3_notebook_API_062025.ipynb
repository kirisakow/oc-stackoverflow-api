```python
import aiohttp
import asyncio
import ast
import pandas as pd


def truncated(string, n_chars, ending='...'):
    return string[:n_chars] + ending if len(string) >= n_chars + len(ending) else string


url_local = 'http://localhost:5555/predict'
url_public = 'http://oc-stackoverflow-api-1-cugua9avgebahha4.francecentral-01.azurewebsites.net'
df_questions = pd.read_csv('data/so.2024.12.csv', converters={'tags': ast.literal_eval})
question_body, expected_tags = df_questions.sample(1, axis=0, ignore_index=True).loc[0, ['body', 'tags']].values

async with aiohttp.ClientSession() as client:
    response = await client.post(url_public, data=question_body)
    predicted_tags = await response.json()

pd.options.display.max_colwidth = None
display(
    pd.DataFrame({
        'Expected tags:': str(sorted(expected_tags)),
        'Predicted tags:': str(sorted(predicted_tags)),
        'Input (title + body, truncated):': truncated(question_body, 300),
    }, index=[0])
)
```
